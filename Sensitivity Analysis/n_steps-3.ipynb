```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker
import numpy as np
df = pd.read_csv(r'E:\A\result\sensitivity_n_steps_results-3.csv')# change to your own path before running

metrics = {
    'Accuracy': ('train_accuracy', 'test_accuracy'),
    'Precision': ('train_precision', 'test_precision'),
    'Recall': ('train_recall', 'test_recall'),
    'F1': ('train_f1', 'test_f1'),
}

n_steps_list = df['n_steps'].astype(str).tolist()
x = np.arange(len(n_steps_list))
width = 0.35

colors_test  = ['#cdc4bb', '#a0a0a2', '#d8b5af', '#68786e']
colors_train = ['#8B7D6B', '#5A5A5C', '#A0625A', '#2E4A38']

fig, axes = plt.subplots(2, 2, figsize=(14, 9))

for ax, (metric, (train_col, test_col)), c_tr, c_te in zip(
        axes.flatten(), metrics.items(), colors_train, colors_test):

    bars1 = ax.bar(x - width/2, df[train_col], width, label='Train', color=c_tr, edgecolor='white', linewidth=0.5)
    bars2 = ax.bar(x + width/2, df[test_col],  width, label='Test',  color=c_te, edgecolor='white', linewidth=0.5)

    for bar in list(bars1) + list(bars2):
        h = bar.get_height()
        ax.text(bar.get_x() + bar.get_width()/2, h + 0.0005,
                f'{h*100:.2f}', ha='center', va='bottom', fontsize=7.5, rotation=0)

    ax.set_title(metric, fontsize=13, fontweight='bold')
    ax.set_xticks(x)
    ax.set_xticklabels([f'N={b}' for b in n_steps_list])
    ax.set_xlabel('N_Steps', fontsize=10)
    ax.set_ylabel('(%)', fontsize=10)
    ax.set_ylim(0.89, 0.99)
    ax.yaxis.set_major_formatter(ticker.FuncFormatter(lambda x, _: f'{x*100:.2f}'))
    ax.legend(fontsize=9)
    ax.spines[['top','right']].set_visible(False)

plt.tight_layout()
plt.savefig(r'E:\A\result\sensitivity_n_steps-33.png', dpi=150, bbox_inches='tight')
plt.show()
```
