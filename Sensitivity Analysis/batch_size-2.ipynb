```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker
import numpy as np

df = pd.read_csv(r'E:\A\result\sensitivity_batch_size_results-2.csv')# change to your own path before running


metrics = {
    'Accuracy': ('train_accuracy', 'test_accuracy'),
    'Precision': ('train_precision', 'test_precision'),
    'Recall': ('train_recall', 'test_recall'),
    'F1': ('train_f1', 'test_f1'),
}

batch_sizes = df['batch_size'].astype(str).tolist()
x = np.arange(len(batch_sizes))
width = 0.35

colors_train = ['#f59790', '#9eaad1', '#f5dbb6', '#dacfe5']
colors_test  = ['#fcd4d1', '#d2d8ed', '#faecd8', '#ede8f4']
colors_line_train = ['#e06b62', '#6e82b8', '#d9a87a', '#b8a0cc']
colors_line_test  = ['#f0a8a4', '#b0bedd', '#f0cfa0', '#cec3df']

fig, axes = plt.subplots(2, 2, figsize=(14, 9))

for ax, (metric, (train_col, test_col)), c_tr, c_te, cl_tr, cl_te in zip(
         axes.flatten(), metrics.items(),
        colors_train, colors_test,
        colors_line_train, colors_line_test):

    bars1 = ax.bar(x - width/2, df[train_col], width, label='Train', color=c_tr, edgecolor='white', linewidth=0.5)
    bars2 = ax.bar(x + width/2, df[test_col],  width, label='Test',  color=c_te, edgecolor='white', linewidth=0.5)

    for bar in list(bars1) + list(bars2):
        h = bar.get_height()
        ax.text(bar.get_x() + bar.get_width()/2, h + 0.0005,
                f'{h*100:.2f}', ha='center', va='bottom', fontsize=7.5, rotation=0)

    ax.plot(x - width/2, df[train_col], color=cl_tr, linewidth=1.8,
            marker='o', markersize=4.5, linestyle='-',  label='Train trend')
    ax.plot(x + width/2, df[test_col],  color=cl_te, linewidth=1.8,
            marker='s', markersize=4.5, linestyle='--', label='Test trend')

    ax.set_title(metric, fontsize=13, fontweight='bold')
    ax.set_xticks(x)
    ax.set_xticklabels([f'BS={b}' for b in batch_sizes])
    ax.set_xlabel('Batch Size', fontsize=10)
    ax.set_ylabel('(%)', fontsize=10)
    ax.set_ylim(0.83, 0.97)
    ax.yaxis.set_major_formatter(ticker.FuncFormatter(lambda x, _: f'{x*100:.2f}'))
    ax.legend(fontsize=9)
    ax.spines[['top','right']].set_visible(False)

plt.tight_layout()
plt.savefig(r'E:\A\result\sensitivity_batch_size-222.png', dpi=150, bbox_inches='tight')
plt.show()
```
